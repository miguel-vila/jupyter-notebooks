```scala211
import $ivy.`dev.zio::zio:1.0.0-RC18-2`, zio._
import $ivy.`dev.zio::zio-logging:0.2.8`, zio.logging._, zio.logging.log._

val env =
    Logging.console(
      format = (_, logEntry) => logEntry,
      rootLoggerName = Some("default-logger")
    )   

def run[R >: Logging, E, A](zio: ZIO[R,E,A]) = Runtime.default.unsafeRun(zio.provideLayer(env))
```

```scala211
run {
    ZIO.fromEither[Exception, Int](Left(new Exception("oh no!")))
      .flatMapError{ err => error(s"My error log: ${err}").as(err) }
}
```

```scala211
run{
    ZIO.fromEither[Exception, Int](Left(new Exception("oh no!")))
      .onError{ err => error(s"My error log: ${err}") }
}
```
