```python
class Node(object):
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.prev_node = None
        self.next_node = None
        
    def set_next(self, next_node):
        self.next_node = next_node

    def set_prev(self, prev_node):
        self.prev_node = prev_node
        
    def get_prev(self):
        return self.prev_node
    
    def get_next(self):
        return self.next_node
    
    def __str__(self):
        prev = 'None' if self.prev_node == None else self.prev_node.key
        next = 'None' if self.next_node == None else self.next_node.key
        return f'Node(key = {self.key}, value = {self.value}, prev = {prev}, next = {next})'

class LRUCache(object):

    def __init__(self, capacity):
        """
        :type capacity: int
        """
        self.capacity = capacity
        self.map = {}
        self.most_recent = None
        self.least_recent = None

    def get(self, key):
        """
        :type key: int
        :rtype: int
        """
        self.debug(f'get {key}')
        if not key in self.map:
            return -1
        node = self.map[key]
        if node == self.least_recent:
            self.least_recent = self.least_recent.get_prev()
        if node.get_prev() == None:
            return node.value
        prev = node.get_prev()
        next = node.get_next()
        prev.set_next( next )
        if next != None:
            next.set_prev(prev)
            
        node.set_prev(None)
        self.most_recent.set_prev( node )
        node.set_next(self.most_recent)
        self.most_recent = node
        return node.value
        
    def debug(self, fn):
        print(f'{fn}')
        print(self.most_recent)
        print(self.least_recent)

    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """
        self.debug(f'put {key} {value}')
        node = Node(key, value)
#         print(self.map.keys())
        if len(self.map) < self.capacity:
            if self.least_recent == None:
                self.least_recent = node
                self.most_recent = node
            else:
                self.most_recent.set_prev( node )
                node.set_next( self.most_recent )
                self.most_recent = node
        else:
            old_least_recent = self.least_recent
#             print(f'DELETING {old_least_recent.key}')
            del self.map[old_least_recent.key]
#             print(f'lest = {self.least_recent}')
            self.least_recent = old_least_recent.get_prev()
            self.least_recent.set_next(None)
            
            self.most_recent.set_prev( node )
            node.set_next( self.most_recent )
            self.most_recent = node
        self.map[key] = node        


# Your LRUCache object will be instantiated and called as such:
c = LRUCache(1)
c.put(2,1)
c.get(2)
c.put(3,2)
```
